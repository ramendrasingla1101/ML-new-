# NUMPY ARRAY


import numpy as np
from matplotlib import pyplot as pit
import pandas as pd
import cv2
import skimage.io as io


```python
lis=[[1,2],[3,4]]
arr1 = np.array(lis, dtype=np.uint8)
arr2 = np.asarray(lis, dtype=np.uint16)

print(lis, type(lis))
print(arr1,type(arr1),arr1.dtype)
```

```python
#New numpy arrays
arr_zero = np.zeros((5,4))   #Argument as tupple(Zero array)
print(arr_zero, arr_zero.shape)
```

```python
#Identity array
arr_i=np.eye(5)
print(arr_i)
arr_i[2][3]=7
print(arr_i)
```

```python
#Copy numpy array
arr_c=arr1.copy()
print(arr_c)
```

```python
# Shapes of numpy arrays and multi dimensional arrays
arr3 = np.random.randint(low=-10,high=10,size=(4,5))
print(arr3)
```

```python
# Reshape 
arr4=arr3.reshape(-1,1)
print(arr4,arr4.shape)
```

```python
# Transpose
print(arr4.T,arr4.T.shape)
```

```python
# Concatenate
arr6 = np.array([[2,4,4],[4,5,6]])
arr7 = np.array([[2,6,9],[3,5,6]])
C= np.concatenate((arr6,arr7),axis=0) #Concatenation by rows
D = np.concatenate((arr6,arr7),axis=1) #Concatenation by coloumns
# print(C)
# print(D)

# Concatenation in expanded dimensions
E = np.concatenate((np.expand_dims(arr6,axis=0), np.expand_dims(arr7,axis=0)),axis=2)
E
```

# MATPLOT-LIB

```python
import matplotlib.pyplot as plt
# plot simple ranges
plt.style.use("fast")
a=np.array(range(0,100,2))
b = np.array(range(0,100,2))

plt.figure(0)
plt.title("Line")
plt.grid("on")
plt.xlabel("year")
plt.xlim(0,100)
plt.ylabel("sales")
plt.ylim(0,100)
plt.plot(a,b,'r',label="2018")
plt.scatter(a,2*a,c='g',label="Nov",marker="*")
plt.legend()
plt.show()
```

```python
# Plot sine wave, cosine wave, and some lines

x=np.linspace(-2*np.pi, 2*np.pi, num=100)
plt.figure(1)
plt.plot(x,np.sin(x),'r')
plt.plot(x,np.cos(x),'g')
plt.show()
```
