```python
# Normal Distributive Function
arr=np.arange(-5,5,0.01)
u=float(input())
sig=float(input())
def gaussian_transform():
    a=0
    fd=[]
    pi=math.pi
    for i in arr:
        a=((i-u)**2)/(2*(sig**2))
        e=math.exp(-a)
        fd.append((1/(sig*math.sqrt(2*pi))) * e)
    return fd
y=gaussian_transform()

plt.plot(arr,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```

```python
# Standard Normal Distributive Function
arr=np.arange(-5,5,0.01)
u=0
sig=1
def gaussian_transform():
    a=0
    fd=[]
    pi=math.pi
    for i in arr:
        a=((i-u)**2)/(2*(sig**2))
        e=math.exp(-a)
        fd.append((1/(sig*math.sqrt(2*pi))) * e)
    return fd
y=gaussian_transform()

plt.plot(arr,y)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```

```python
# Random Scatter Plot
def frequency(a,b):
    lis=[]
    c=0
    for x in a:
        c=0
        for y in b:
            if (x==y):
                c=c+1
        
        lis.append(c)
    
    arr=np.array(lis, dtype=np.uint16)
    return arr

a=np.array(range(0,100))
b=np.random.randint(low=0,high=100,size=(1000))
freq=frequency(a,b)

plt.figure(1)
plt.title("Random Numbers Frequency")
plt.grid("on")
plt.xlabel("X")
plt.xlim(0,100)
plt.ylabel("Frequency")
plt.ylim(0,50)
plt.scatter(a,freq,c='r')
plt.show()
```
