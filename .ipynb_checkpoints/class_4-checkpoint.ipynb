# By Gradient Descent

```python
def hypothesis(x,theta):
    return x*theta[1]+ theta[0]

def prediction(X,theta):
    y_pred = []
    for i in range(X.shape[0]):
        pred = hypothesis(X[i],theta)
        y_pred.append(pred)
    y_pred = np.array(y_pred)
    return y_pred

def get_error(X,Y,theta):
        e = 0
        m = X.shape[0]
        for i in range(m):
            e+=(hypothesis(X[i],theta)-Y[i])**2
        return e

def getGradients(X,Y,theta):
    grads = np.zeros((2))
    m = X.shape[0]
    
    for i in range(m):
        YPi = hypothesis(X[i],theta)
        grads[0]+= (YPi - Y[i])
        grads[1]+= (YPi - Y[i])*X[i]
        
    return grads

def gradientDescent(X,Y,maxItr=100,learning_rate=0.001):
    theta = np.zeros((2))
    error = []
    
    for i in range(maxItr):
        grad = getGradients(X,Y,theta)
        e = get_error(X,Y,theta)
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        error.append(e)
        
    return theta,error
```

```python
theta, error = gradientDescent(X,Y)
```

```python
plt.figure(1)
plt.scatter(X,Y)
plt.plot(X,prediction(X,theta),c='red')
plt.show()
```

# By Closed Form Solution

```python

A = np.ones((X.shape[0],1))
X = np.hstack((A,X))
print(X.shape)
Y = Y.reshape((-1,1))
theta = np.dot(np.linalg.pinv(np.dot(X.T,X)),np.dot(X.T,Y))
theta = theta.reshape((-1,))
print(theta)
```

```python
x_test = X
y_test = []
for i in range(x_test.shape[0]):
    y_test.append(hypothesis(x_test[i,1],theta))
y_test = np.array(y_test)

plt.scatter(X[:,1:],Y)
plt.plot(x_test[:,1:],y_test,c='red')
plt.show()
```
