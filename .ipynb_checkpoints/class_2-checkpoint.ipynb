# KNN OF NORMAL DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt
mean1=np.array([0.0,0.0])
mean2=np.array([5.0,7.0])
cov1=np.array([[.5, 0],[0,.5]])
cov2=np.array([[2.5,-1.8],[-1.8,3.0]])
data1 = np.random.multivariate_normal(mean1,cov1,500)
data2 = np.random.multivariate_normal(mean2,cov2,500)
plt.figure(0)
plt.scatter(data1[:,0],data1[:,1],c='gray')
plt.scatter(data2[:,0],data2[:,1],c='black')
plt.show()
    
    
```

```python
query_x= np.array([4,4])
plt.scatter(data1[:,0],data1[:,1],c='gray')
plt.scatter(data2[:,0],data2[:,1],c='black')
plt.scatter(query_x[0],query_x[1],color='red')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

# Test Time
def knn(data1,data2,queryPoint,k=5):
    vals=[]
    m=data1.shape[0]
    n=data2.shape[0]
    
    for i in range(m):
        d=dist(queryPoint,data1[i])
        vals.append((d,0))
    for i in range(n):
        d=dist(queryPoint,data2[i])
        vals.append((d,1))
    
    vals = sorted(vals)
    # Nearest/First K points
    vals = vals[:k]
    vals = np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
```

```python
x=knn(data1,data2,query_x)
print(x)
```

# KNN using sklearn library

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
X, Y = make_blobs(n_samples=400, n_features=2,center_box=(-2,2),centers=2,shuffle=True,random_state=11)
```

```python
print(X[:5,:])
```

```python
print(Y[:5])
```

```python
query_x = np.array([0.2,0.0])


plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.scatter(query_x[0],query_x[1],c='red')
plt.title("KNN Classification")
plt.xlabel("x1 feature")
plt.ylabel("x2 feature")
plt.show()
```

```python
def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))

def knn(X,Y,query_x,k=5):
    m=X.shape[0]
    d_list=[]
    for i in range(m):
        d=distance(X[i],query_x)
        d_list.append((d,Y[i]))
    
    d_list = sorted(d_list,reverse=False)
    d_list = np.array(d_list[:k])
    d_list = d_list[:,1]
    freq = np.unique(d_list,return_counts=True)
    idx = np.argmax(freq[1])
    pred = freq[0][idx]
    
    return int(pred)

knn(X,Y,query_x)
```

```python
# ACCURACY
def predict(X,Y):
    
    all_pred = []
    
    for x in X:
        pred = knn(X,Y,x)
        all_pred.append(pred)
    
    
    all_pred = np.array(all_pred,dtype='int32')
    #print(all_pred)
    return all_pred

pred = predict(X,Y)
print((np.sum(Y==pred)/Y.shape[0]))
```

# OPEN-CV Prerequisites for Face Detection Model

```python
import cv2
import numpy as np
```

```python
img = cv2.imread("/home/ramendra/Desktop/ML/img.jpg")
img = cv2.resize(img,(100,100))
print(img.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.show()
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
def transform(img,orig_rgb,new_rgb):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if((img[i,j,0],img[i,j,1],img[i,j,2])==orig_rgb):
                img[i,j,0],img[i,j,1],img[i,j,2]=new_rgb
        
    return img
img = transform(img,(255,255,255),(0,0,0))
plt.imshow(img)
plt.show()
```
