## Simple Convolution

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('/home/ramendra/Desktop/ML-new-/Neural Networks/759351.jpg')
img = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)

print (img.shape)
plt.imshow(img, cmap='gray')
```

```python
kernel = np.array([[-1.0, -1.0, -1.0], [-1.0,8.0,-1.0], [-1.0,-1.0,-1.0]])

print(kernel)
```

```python
new_img = np.zeros((img.shape[0]-kernel.shape[0]+1,img.shape[1]-kernel.shape[1]+1))
```

```python
# convolution

for ix in range(new_img.shape[0]):
    for ij in range(new_img.shape[1]):
        
        #patch
        img_patch = img[ix:ix+kernel.shape[0], ij:ij+kernel.shape[1]]
        k_prod = img_patch*kernel
        
        new_img[ix][ij] = max(0,k_prod.sum())
```

```python
plt.imshow(new_img.astype(np.uint8), cmap='gray')
```

```python
X_train = X.reshape((10000,28,28,1))

# one-hot encoding
Y_train = np_utils.to_categorical(y)

print(X_train.shape, Y_train.shape)
```

```python
model = Sequential()
model.add(Convolution2D(32,(3,3),activation='relu', input_shape=(28,28,1)))
model.add(Convolution2D(64, (3,3), activation='relu'))
model.add(Dropout(0.25))
model.add(MaxPooling2D(2,2))
model.add(Convolution2D(32,(5,5),activation='relu'))
model.add(Convolution2D(8, (5,5),activation='relu'))
model.add(Flatten())
model.add(Dense(10,activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
hist = model.fit(X_train,Y_train,
                 epochs = 10,
                 shuffle = True,
                 batch_size = 256,
                 validation_split=0.20)
```

```python
plt.figure(0)
plt.plot(hist.history['loss'],'g')
plt.plot(hist.history['val_loss'],'b')
plt.plot(hist.history['acc'],'r')
plt.plot(hist.history['val_acc'],'black')
plt.show()
```
